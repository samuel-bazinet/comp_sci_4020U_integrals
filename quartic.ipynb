Let 
$$
    \int_{a}^{b}f(x)dx = A_1f(a) + A_2f(m) + A_3f(n) + A_4f(o) + A_5f(b)
$$

then

$$
    \int_{a}^{b}1dx = b-a = A_1 + A_2 + A_3 + A_4 + A_5
$$

$$
    \int_{a}^{b}xdx = \frac{b^2-a^2}{2} = A_1{a} + A_2(a+\frac{b-a}{4}) + A_3{(a+2\frac{b-a}{4})} + A_4{(a+3\frac{b-a}{4})} + A_5{b}
$$

$$
    \int_{a}^{b}x^2dx = \frac{b^3-a^3}{3} = A_1{a^2} + A_2(a+\frac{b-a}{4})^2+ A_3{(a+2\frac{b-a}{4})}^2 + A_4{(a+3\frac{b-a}{4})}^2 + A_4{b^2}
$$

$$
    \int_{a}^{b}x^3dx = \frac{b^4-a^4}{4} = A_1{a^3} + A_2(a+\frac{b-a}{4})^3+ A_3{(a+2\frac{b-a}{4})}^3 + A_3{(a+3\frac{b-a}{4})}^3 + A_4{b^3}
$$


$$
    \int_{a}^{b}x^4dx = \frac{b^5-a^5}{5} = A_1{a^4} + A_2(a+\frac{b-a}{4})^4+ A_3{(a+2\frac{b-a}{4})}^4 + A_3{(a+3\frac{b-a}{4})}^4 + A_4{b^4}
$$

Using Guass-Jordan eimination

$$
\left[
\begin{array}{ccccc|c}
A_1 & A_2 & A_3 & A_4 & A_5 & b-a\\
A_1a & A_2(a +\frac{b-a}{4}) & A_3(a+2\frac{b-a}{4}) & A_4(a+3\frac{b-a}{4}) & A_5b & \frac{b^2-a^2}{2}\\
A_1a^2 & A_2(a +\frac{b-a}{4})^2 & A_3(a+2\frac{b-a}{4})^2 & A_4(a+3\frac{b-a}{4})^2 & A_5b^2 & \frac{b^3-a^3}{3}\\
A_1a^3 & A_2(a +\frac{b-a}{4})^3 & A_3(a+2\frac{b-a}{4})^3 & A_4(a+3\frac{b-a}{4})^3 & A_5b^3 & \frac{b^4-a^4}{4}\\
A_1a^4 & A_2(a +\frac{b-a}{4})^4 & A_3(a+2\frac{b-a}{4})^4 & A_4(a+3\frac{b-a}{4})^4 & A_5b^4 & \frac{b^5-a^5}{5}
\end{array}
\right]
$$

Using Maple to solve this, we get:

$A_1 = 7\frac{b-a}{90}$

$A_2 = 32\frac{b-a}{90}$

$A_3 = 12\frac{b-a}{90}$

$A_4 = 32\frac{b-a}{90}$

$A_5 = 7\frac{b-a}{90}$

Using those coefficents, we can make the cubic rule:

$$
\int_{a}^{b} f(x)dx \approx \frac{1}{90} \left[7f(a) + 32f(a+\frac{b-a}{4}) + 12f(a + 2\frac{b-a}{4}) + 32f(a + 3\frac{b-a}{4}) + 7f(b)\right](b-a)
$$

```python
import numpy as np
from typing import Callable

def quartic(f: Callable[[float], float], i: int, a: float, b: float):
    xs = np.linspace(a, b, i, True, dtype = np.float64)
    A: list[np.float64] = []
    assert i >=2
    for i in range(1, len(xs)):
        m_1 = (xs[i] + (xs[i-1]-xs[i])/4)
        m_2 = (xs[i] + 2*(xs[i-1]-xs[i])/4)
        m_3 = (xs[i] + 3*(xs[i-1]-xs[i])/4)
        a = (1/90) * (7*f(xs[i]) + 32*f(m_1) + 12*f(m_2) + 32*f(m_3) + 7*f(xs[i-1]))*(xs[i]-xs[i-1])
        A.append(a)
    return sum(A)
```
